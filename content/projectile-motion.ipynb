## Projectile Motion Predictions

### What's a Linear Regression?

A Linear Regression (best-fit line) is a technique for approximating a function from data points. We'll use a linear regression to predict a distance from pressure (in PSI) from a pressurized projectile launcher.

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

# https://realpython.com/linear-regression-in-python/#simple-linear-regression-with-scikit-learn
def predict_linear_regression(x: list[float], y: list[float], xlabel: str, ylabel: str, test_x: float):
    model = LinearRegression()
    # the reshape function transforms the data to be "horizontal"
    x = np.array(x).reshape((-1, 1))
    y = np.array(y)
    model.fit(x, y)
    b = model.intercept_
    m = model.coef_[0]

    # plot the linear regression
    linreg_x = np.linspace(0, 150, num=50)
    plt.plot(linreg_x, m*linreg_x + b)
    # plot the data points
    plt.scatter(x, y, color='r')
    # plot the predicted point
    predict_y = m*test_x + b
    plt.scatter([test_x], [predict_y], color='g')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend([f'Linear Regression\ny = {m:.3f} * x + {b:.3f}', "data", "predicted point"])
    plt.show()

    return predict_y
```

### Your data

Replace `psi`, `distance`, and `test_psi` with your own values. Under the plot is the predicted distance value.

```python
psi = [50, 75, 90]
distance = [6.5, 9, 11]
test_psi = 110

predict_linear_regression(psi, distance, "Pressure (PSI)", "Distance (meters)", test_psi)
```
